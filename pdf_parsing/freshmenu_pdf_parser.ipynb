```python
from tika import parser
from parse import parse
import pprint

# file_name = "30232804_FM-KA-6773739.pdf"
file_name = "10720938_FM-KA-2287171"


parsed = parser.from_file(file_name)

splitted = parsed["content"].split("\n")

clean_lines = [line for line in splitted if len(line) != 0]

filter_list = ["Invoice No: {invoice_no}",
               "Total Price: {total_price}",
               "Net Total: {net_total}"]

def do_simple_parsing(clean_lines, filter_list):
    order_details = {}
    for text_filter in filter_list:
        for line in clean_lines:
            result = parse(text_filter, line)
            if result:
                order_details.update(result.__dict__["named"])
    return order_details

def get_element_in_between(starts_with, ends_with, clean_lines):
    for line in clean_lines:
        if starts_with == line:
            start_index = clean_lines.index(line) + 1
        elif ends_with == line:
            end_index = clean_lines.index(line)
        else:
            pass
    element_list = clean_lines[start_index:end_index]
    return element_list

def get_customer_details(clean_lines):
    starts_with = "Name: Phone No: Address:"
    ends_with = "ORDER DETAILS"
    
    customer_details = {}
    
    address_list = get_element_in_between(starts_with, ends_with, clean_lines)
    customer_details["customer_name"] = address_list[0]
    customer_details["customer_address"] = " ".join(address_list[1:])
    
    return customer_details

def get_order_details(clean_lines):
    starts_with = "ORDER DETAILS"
    ends_with = "Payment Status:"
    
    order_details = {}
    
    order_list = get_element_in_between(starts_with, ends_with, clean_lines)
    order_details["order_head"] = order_list[0]
    order_details["order_body"] = order_list[1]
    order_head_splitted = order_details["order_head"].split(":")
    clean_order__head_splitted = [line for line in order_head_splitted if len(line) != 0]
    order_body_splitted = order_details["order_body"].split(" ")
    final_order_details = dict(zip(order_head_splitted, order_body_splitted))
    
    return final_order_details

def get_payment_details(clean_lines):
    starts_with = "Payment Status:"
    ends_with = "ITEM DETAILS"
    
    payment_details = {}
    
    payment_list = get_element_in_between(starts_with, ends_with, clean_lines)
    payment_details["payment_head"] = payment_list[0]
    payment_details["payment_body"] = payment_list[1]

    payment_head_splitted = payment_details["payment_head"].split(": ")
    payment_body_splitted = payment_details["payment_body"].split(" ", 1)
    result = parse("{%te} {%te} {delivered_by}", payment_body_splitted[1])

#     payment_body_splitted = payment_details["payment_body"].split(" ", -1)
    
#     payment_body_splitted2 = payment_body_splitted[1].split(" ")
#     final_order_details = dict(zip(order_head_splitted, order_body_splitted))
    
#     return final_order_details
    return payment_body_splitted
#     return result

    

# print(do_simple_parsing(clean_lines, filter_list))

# print(get_customer_details(clean_lines))
# print(get_order_details(clean_lines))
print(get_payment_details(clean_lines))
```
